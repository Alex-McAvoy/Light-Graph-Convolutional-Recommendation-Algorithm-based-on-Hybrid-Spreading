```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
movielens = pd.read_excel("/data/alex/algorithm/movielens/evaluation/model_evaluation_results.xlsx", sheet_name=1)
douban = pd.read_excel("/data/alex/algorithm/douban/evaluation/model_evaluation_results.xlsx", sheet_name=1)

movielens = movielens[movielens["Model"].isin(["SpreadLightGCN", "SpreadLightGCNOpti"])]
douban = douban[douban["Model"].isin(["SpreadLightGCN", "SpreadLightGCNOpti"])]


movielens_model = ["LGCNHS-e", "LGCNHS"]
movielens_P = movielens.P
movielens_R = movielens.R
movielens_F1 = movielens.F1
movielens_NDCG = movielens.NDCG
movielens_H = movielens.H
movielens_I = movielens.I
douban_model = ["LGCNHS-e", "LGCNHS"]
douban_P = douban.P
douban_R = douban.R
douban_F1 = douban.F1
douban_NDCG = douban.NDCG
douban_H = douban.H
douban_I = douban.I
```

```python
def annotate_bars(ax, values):
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.4f}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5), 
                    textcoords='offset points')

sns.set(style="whitegrid")
sns.set(style="ticks")

plt.figure(figsize=(8, 6))

plt.subplot(2,3,1)
ax1 = sns.barplot(x=movielens_model, y=movielens_P, palette="viridis")
plt.title("(a)P@30 Comparison")
plt.ylabel("P@30")
plt.xlabel("Models")
plt.ylim(0.092, 0.106)
# annotate_bars(ax1, movielens_P)
# plt.xticks(rotation=45)

plt.subplot(2,3,2)
ax2 = sns.barplot(x=movielens_model, y=movielens_R, palette="viridis")
plt.title("(b)R@30 Comparison")
plt.ylabel("R@30")
plt.xlabel("Models")
plt.ylim(0.5, 0.55)
# annotate_bars(ax2, movielens_R)
# plt.xticks(rotation=45)

plt.subplot(2,3,3)
ax3 = sns.barplot(x=movielens_model, y=movielens_F1, palette="viridis")
plt.title("(c)F1@30 Comparison")
plt.ylabel("F1@30")
plt.xlabel("Models")
plt.ylim(0.15, 0.18)
# annotate_bars(ax3, movielens_F1)
# plt.xticks(rotation=45)

plt.subplot(2,3,4)
ax4 = sns.barplot(x=movielens_model, y=movielens_NDCG, palette="viridis")
plt.title("(d)NDCG@30 Comparison")
plt.ylabel("NDCG@30")
plt.xlabel("Models")
plt.ylim(0.12, 0.136)
# annotate_bars(ax4, movielens_NDCG)
# plt.xticks(rotation=45)

plt.subplot(2,3,5)
ax5 = sns.barplot(x=movielens_model, y=movielens_H, palette="viridis")
plt.title("(e)H@30 Comparison")
plt.ylabel("H@30")
plt.xlabel("Models")
plt.ylim(0.7, 0.85)
# annotate_bars(ax5, movielens_H)
# plt.xticks(rotation=45)

plt.subplot(2,3,6)
ax6 = sns.barplot(x=movielens_model, y=movielens_I, palette="viridis")
plt.title("(f)I@30 Comparison")
plt.ylabel("I@30")
plt.xlabel("Models")
plt.ylim(0, 0.6)
# annotate_bars(ax6, movielens_I)
# plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
