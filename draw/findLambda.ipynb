```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
movielens = pd.read_csv("/data/alex/algorithm/movielens/evaluation/lambda_evaluation_30.csv")
douban = pd.read_csv("/data/alex/algorithm/douban/evaluation/lambda_evaluation_30.csv")

movielens_lambda_ls = movielens["lambda"]
movielens_P_ls = movielens["precision"]
movielens_R_ls = movielens["recall"]
movielens_F1_ls = movielens["f1"]
movielens_NDCG_ls = movielens["ndcg"]
movielens_H_ls = movielens["H"]
movielens_I_ls = movielens["I"]

douban_lambda_ls = douban["lambda"]
douban_P_ls = douban["precision"]
douban_R_ls = douban["recall"]
douban_F1_ls = douban["f1"]
douban_NDCG_ls = douban["ndcg"]
douban_H_ls = douban["H"]
douban_I_ls = douban["I"]
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(15,6))

plt.subplot(2,6,1)
sns.lineplot(x=movielens_lambda_ls, y=movielens_P_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("P")
plt.title("(a)Movielens P@30")
# plt.text(x=0.5, y=-0.1, s="Movielens P@30", ha='center', transform=plt.gca().transAxes)
plt.subplot(2,6,2)
sns.lineplot(x=movielens_lambda_ls, y=movielens_R_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("R")
plt.title("(b)Movielens R@30")
# plt.text(x=0.5, y=-0.1, s="Movielens R@30", ha='center', transform=plt.gca().transAxes)

plt.subplot(2,6,3)
sns.lineplot(x=movielens_lambda_ls, y=movielens_F1_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("F1")
plt.title("(c)Movielens F1@30")
# plt.text(x=0.5, y=-0.1, s="Movielens F1@30", ha='center', transform=plt.gca().transAxes)

plt.subplot(2,6,4)
sns.lineplot(x=movielens_lambda_ls, y=movielens_NDCG_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("NDCG")
plt.title("(d)Movielens NDCG@30")
# plt.text(x=0.5, y=-0.1, s="Movielens NDCG@30", ha='center', transform=plt.gca().transAxes)

plt.subplot(2,6,5)
sns.lineplot(x=movielens_lambda_ls, y=movielens_H_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("H")
plt.title("(e)Movielens H@30")
# plt.text(x=0.5, y=-0.1, s="Movielens H@30", ha='center', transform=plt.gca().transAxes)


plt.subplot(2,6,6)
sns.lineplot(x=douban_lambda_ls, y=movielens_I_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("I")
plt.title("(f)Movielens I@30")
# plt.text(x=0.5, y=-0.1, s="Movielens I@30", ha='center', transform=plt.gca().transAxes)


plt.subplot(2,6,7)
sns.lineplot(x=douban_lambda_ls, y=douban_P_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("P")
plt.title("(g)Douban P@30")
# plt.text(x=0.5, y=-0.1, s="Douban P@30", ha='center', transform=plt.gca().transAxes)

plt.subplot(2,6,8)
sns.lineplot(x=douban_lambda_ls, y=douban_R_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("R")
plt.title("(h)Douban R@30")
# plt.text(x=0.5, y=-0.1, s="Douban R@30", ha='center', transform=plt.gca().transAxes)

plt.subplot(2,6,9)
sns.lineplot(x=douban_lambda_ls, y=douban_F1_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("F1")
plt.title("(i)Douban F1@30")
# plt.text(x=0.5, y=-0.1, s="Douban F1@30", ha='center', transform=plt.gca().transAxes)

plt.subplot(2,6,10)
sns.lineplot(x=douban_lambda_ls, y=douban_NDCG_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("NDCG")
plt.title("(j)Douban NDCG@30")
# plt.text(x=0.5, y=-0.1, s="Douban NDCG@30", ha='center', transform=plt.gca().transAxes)

plt.subplot(2,6,11)
sns.lineplot(x=douban_lambda_ls, y=douban_H_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("H")
plt.title("(k)Douban H@30")
# plt.text(x=0.5, y=-0.1, s="Douban H@30", ha='center', transform=plt.gca().transAxes)

plt.subplot(2,6,12)
sns.lineplot(x=douban_lambda_ls, y=douban_I_ls, color="blue")
plt.xlabel("lambda")
plt.ylabel("I")
plt.title("(l)Douban I@30")
# plt.text(x=0.5, y=-0.1, s="Douban I@30", ha='center', transform=plt.gca().transAxes)


plt.tight_layout()
plt.show()
```
